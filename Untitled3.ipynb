```python
from Site.models import Site
```

```python
sites = Site.objects.all()
with open('site_counter_for_kuo.jpg','r') as f1:
    sites_counter = eval(f1.read())
    print(sites_counter)
    for siteCounter in sites_counter:
        print(siteCounter)
        for site in sites:
            if site.name == siteCounter:
                if 'b' in site_counter[siteCounter]:
                    site.boyAsset = sites_counter[siteCounter]['b']
                else:
                    site.boyAsset = 0
                if 'g' in site_counter[siteCounter]:
                    site.girlAsset = sites_counter[siteCounter]['g']
                else:
                    site.girlAsset = 0
                if '同學' in site_counter[siteCounter]:
                    site.classmateAsset = sites_counter[siteCounter]['同學']
                else:
                    site.classmateAsset = 0
                if '親子' in site_counter[siteCounter]:
                    site.childrenAsset = sites_counter[siteCounter]['親子']
                else:
                    site.childrenAsset = 0
                if '家庭' in site_counter[siteCounter]:
                    site.familyAsset  = sites_counter[siteCounter]['家庭']
                else:
                    site.familyAsset = 0
                if '情侶' in site_counter[siteCounter]:
                    site.coupleAsset = sites_counter[siteCounter]['情侶']
                else:
                    site.coupleAsset = 0
                site.save()
                break
```
