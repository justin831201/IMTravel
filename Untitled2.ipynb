```python
from pymongo import MongoClient
from Site.models import Site
from Country.models import Region
client = MongoClient(port=27017)
```

```python
db = client["imtravel"]
scoreTable = db["台北score"]
scores = list(scoreTable.find({}))
keywordTable = db["台北keyword"]
keywords = list(keywordTable.find({}))
```

```python
taipei = Region.objects.get(chinese_name="台北")
oldSites = Site.objects.filter(isOpen=True, region=taipei)
```

```python
for site in oldSites:
    site.score = None
    site.keyword = None
    for score in scores:
        if score['site'] == site.name:
            if score['score'] != 0:
                site.score = round((score['score']-2.0)*5.0/3.0) +1
    for keyword in keywords:
        if keyword['site'] == site.name:
            site.keyword = keyword['keywords'][:3]
    if site.score != None and site.keyword != None:
        print(site.name + " " + str(site.score))
        print(site.keyword)
        print("------------------")
sites = []
```

```python
for site in oldSites:
    sites.append(site)
```
