```python
data = request.json()
```

```python
results = data["Infos"]["Info"]
for result in results:
    print("名稱："+result["Name"])
    print("經度："+result["Px"])
    print("緯度："+result["Py"])
    (name, city) = region(int(result["Zipcode"]))
    City = Region.objects.get(chinese_name = name)
    print("城市："+str(City))
    print("地址："+city + "  "+result["Add"])
    print("描述："+result["Toldescribe"])
    print("交通："+result["Travellinginfo"])
    print("時間："+result["Opentime"])
    name1 = category(int(result["Class1"]))
    Cat = Category.objects.get(chinese_name = name1)
    print("分類："+str(Cat))
    if result["Picture1"] != "":
        print("相關照片:\n"+ result["Picture1"])
        print(result["Picdescribe1"])
        if result["Picture2"] != "":
            print(result["Picture2"])
            print(result["Picdescribe2"])
        if result["Picture3"] !="":
            print(result["Picture3"])
            print(result["Picdescribe3"])
    print("")
```
